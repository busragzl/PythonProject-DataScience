```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
df[["sex","class","embark_town"]].isna().sum()
```

```python
df = df.dropna(axis=0, subset=(["embark_town"]))
```

```python
df
```

```python
df[["sex","class","embark_town"]].isna().sum()
```

```python
#one hot encoding
```

```python
df["sex"].value_counts()
```

```python
df["class"].value_counts()
```

```python
df["embark_town"].value_counts()
```

```python
df_onehot = pd.get_dummies(df,columns=["sex","embark_town"],drop_first=True)
```

```python
df_onehot
```

```python
df_onehot.columns
```

```python
# labelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
df_label = df.copy()
```

```python
df_label
```

```python
df_label["sex"]= label_encoder.fit_transform(df_label["sex"])
```

```python
df_label
```

```python
#ordinal encoder
```

```python
df_ordinal = df.copy()
```

```python
class_order = ["Third","Second","First"]
```

```python
ordinal_encoder = OrdinalEncoder(categories = [class_order])
```

```python
df_ordinal["class"] = ordinal_encoder.fit_transform(df_ordinal[["class"]])
```

```python
df_ordinal
```

```python
fig , axis = plt.subplots(1,3,figsize=(15,5))
df["sex"].value_counts().plot(kind="bar",ax = axis[0], title = "Ordinal Categories")
df_label["sex"].value_counts().plot(kind="bar",ax = axis[1], title = "Label Encoding")
df_onehot["sex_male"].value_counts().plot(kind="bar",ax = axis[2], title = "One hot Encoding")
```
