```python
import numpy as np
```

```python
import pandas as pd
```

```python
# random seed

np.random.seed(42) # tutalı rastgelelik elde etmek için

set1no = 900
set2no = 100
```

```python
df1 = pd.DataFrame({

    "feature_1" : np.random.normal(loc=0,scale=1,size=set1no), # normal dağılıma göre
    "feature_2" : np.random.normal(loc=0,scale=1,size=set1no),
    "target" : [0] * set1no
})

df2 = pd.DataFrame({

    "feature_1" : np.random.normal(loc=0,scale=1,size=set2no), # normal dağılıma göre
    "feature_2" : np.random.normal(loc=0,scale=1,size=set2no),
    "target" : [1] * set2no
})
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.concat([df1,df2]).reset_index(drop=True)
```

```python
df
```

```python
df["target"].unique()
```

```python
df["target"].value_counts()
```

```python
#upsampling -> upsample minority -> azınlık olanları arttırmak 
```

```python
#downsampling -> downsample majority -> çoğunluğu azaltmak 
```

```python
#upsampling
```

```python
df_minority = df[df["target"] == 1]
```

```python
df_minority
```

```python
df_majority = df[df["target"] == 0]
```

```python
df_majority
```

```python
from sklearn.utils import resample
```

```python
df_minority_upsampled=resample(df_minority,replace=True,n_samples=len(df_majority), random_state=42)
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled = pd.concat([df_majority,df_minority_upsampled])
```

```python
df_upsampled["target"].value_counts()
```

```python
#downsampling
```

```python
df_majority_downsampled=resample(df_majority,replace=True,n_samples=len(df_minority), random_state=42)
```

```python
df_majority_downsampled
```

```python
df_majority_downsampled["target"].value_counts()
```

```python
df_downsampled = pd.concat([df_minority,df_majority_downsampled])
```

```python
df_downsampled["target"].value_counts()
```

```python
# SMOTE (Synthetic Mimority Over-sampling Technique) -> Benzer data oluşturarak upsampling
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df["feature_1"],df["feature_2"],c=df["target"])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
```

```python
(X,y) = oversample.fit_resample(df[["feature_1","feature_2"]], df["target"]) #-> y genelde target olur diğerleri X feature'lar
```

```python
X
```

```python
y
```

```python
oversample_df = pd.concat([X,y],axis=1)
```

```python
oversample_df["target"].value_counts()
```

```python
plt.scatter(oversample_df["feature_1"],oversample_df["feature_2"],c=oversample_df["target"])
```
