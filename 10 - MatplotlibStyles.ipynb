```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1,data2,"g", alpha = 0.5) # alpha -> transparent 
my_axes.plot(data2,data1 , color = '#1E0703')
plt.show()
```

```python
(new_fig,new_axes) = plt.subplots()
new_axes.plot(data1, data1 + 2 , color = "blue" , linewidth=1.0)
new_axes.plot(data1, data1 + 3 , color = "yellow" , linewidth=2.0)
new_axes.plot(data1, data1 + 4 , color = "green" , linestyle = "-.")
new_axes.plot(data1, data1 + 5 , color = "red" , linestyle = ":")
new_axes.plot(data1, data1 + 6 , color = "purple" , linestyle = "--")
new_axes.plot(data1, data1 + 7 , color = "#000000" , linestyle = "--", marker="o",markersize=8,markerfacecolor="red")
new_axes.plot(data1, data1 + 8 , color = "#000000" , linestyle = "--", marker="+",markersize=10)


plt.show()
```
